###### For Resort Hotel

```python
Resort_hotel=df[df['hotel']=='Resort Hotel']
Resort_hotel=Resort_hotel[Resort_hotel['is_canceled']==0]
print("resort hotel reversation when booking was not cancel")
Resort_hotel.shape
```

##### Hotel Reversation based on the arrival date Month

```python
Resort_hotel=Resort_hotel.groupby(['arrival_date_month'])['adr'].mean().reset_index()
Resort_hotel
```

```python
City_hotel=City_hotel.groupby(['arrival_date_month'])['adr'].mean().reset_index()
City_hotel
```

```python
City_n_resort = Resort_hotel.merge(City_hotel,on='arrival_date_month')
City_n_resort.columns=['Month','Resort Hotel Price','City Hotel Price']
City_n_resort
```

```python
import sort_dataframeby_monthorweek as sd

def sort_month(df, column_name):
    return sd.Sort_Dataframeby_Month(df, column_name)
Month_in_Order = sort_month(City_n_resort, 'Month')
Month_in_Order
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=Month_in_Order,x='Month',y='Resort Hotel Price',color='darkblue',label='Resort Hotel Price')
sns.lineplot(data=Month_in_Order,x='Month',y='City Hotel Price',color='red',label='City Hotel Price')
plt.xlabel("Month",fontsize=15)
plt.ylabel("Value of Resort and City Hotel Price",fontsize=12)
plt.title("Prices Of Hotels According To The Months",fontsize=20)
plt.legend()
plt.grid()
```

- So the figure shows that the Price of Resort hotel are higher during the month of August
- the Price of City Hotels are higher in The Month of the April to may

# Model Building

## 1.kNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
knn= KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
prediction = knn.predict(X_test)
prediction
```

```python
accuracy = accuracy_score(y_test,prediction)
accuracy
```

```python
TESTING_PREDICTIONS=[]
TRAINING_PREDICTIONS=[]
for i in range(3,8,1):
    knn=KNeighborsClassifier(n_neighbors=i, metric="euclidean")
    knn.fit(X_train,y_train)
    
    testing_predictions=knn.predict(X_test)
    test_accuracy=accuracy_score(y_test,testing_predictions)
    TESTING_PREDICTIONS.append(test_accuracy)
    
    traning_predictions=knn.predict(X_train)
    train_accuracy=accuracy_score(y_train,traning_predictions)
    TRAINING_PREDICTIONS.append(train_accuracy)
```

```python
print('Testing Prediction ',TESTING_PREDICTIONS)
print("Training Prediction ",TRAINING_PREDICTIONS)
```

```python
knn=KNeighborsClassifier(n_neighbors=6, metric="euclidean")
knn.fit(X_train,y_train)
    
testing_predictions=knn.predict(X_test)
test_accuracy_knn=accuracy_score(y_test,testing_predictions)
print(test_accuracy_knn)

traning_predictions=knn.predict(X_train)
train_accuracy=accuracy_score(y_train,traning_predictions)
print(train_accuracy)
```

```python
from sklearn.metrics import confusion_matrix
print(" Confusion Matrix \n",confusion_matrix(y_test,testing_predictions))
```

```python
from sklearn.metrics import classification_report
print("Classification Report :\n\n",classification_report(y_test,testing_predictions))
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree=DecisionTreeClassifier()
decision_tree.fit(X_train,y_train)
y_prediction_dec = decision_tree.predict(X_test)
y_prediction_dec
```

```python
accuracy_dec=accuracy_score(y_test,y_prediction_dec)
accuracy_dec
```

```python
max_depth_range=list(range(4,24,2))
ACCURICIES=[]
for i in max_depth_range:
    dt_clfn = DecisionTreeClassifier(max_depth=i,random_state=3)
    dt_clfn.fit(X_train,y_train)
    npredictions= dt_clfn.predict(X_test)
    ACCURICIES.append(accuracy_score(y_test,npredictions))
```

```python
print(ACCURICIES)
```

```python
plt.plot(max_depth_range,ACCURICIES, marker='o',color='darkblue')
plt.xlabel('max_depth')
plt.ylabel('Accuries')
plt.title("ACCURIES BASED OF MAX_DEPTH")
```

```python
decision_tree=DecisionTreeClassifier(max_depth=18,random_state=8)
decision_tree.fit(X_train,y_train)
y_prediction_dec = decision_tree.predict(X_test)
```

```python
accuracy_dec=accuracy_score(y_test,y_prediction_dec)
accuracy_dec
```

```python
confusion_matrix_dec = confusion_matrix(y_test,y_prediction_dec)
confusion_matrix_dec
```

```python
print(classification_report(y_test,y_prediction_dec))
```
